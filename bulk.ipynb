```python
from elasticsearch import Elasticsearch, helpers
from elasticsearch_dsl import Index
import json, re
import codecs
import unicodedata
# import queries

client = Elasticsearch(HOST="http://localhost", PORT=9200)
INDEX = 'scientists_db'

# Creating index if not manually created
# def createIndex():
#     index = Index(INDEX, using=client)
#     res = index.create()
#     print(res)

def read_all_scientists():
    with open("C:/Users/P.Thukaraka/Downloads/codebeautify.json", 'r', encoding='utf-8-sig') as f:
        all_scientists = json.loads("[" +
                          f.read().replace("}\n{", "},\n{") +"]")
        # all_scientists = json.loads(f.read())
        res_list = [i for n, i in enumerate(all_scientists) if i not in all_scientists[n + 1:]]
        return res_list

def genData(scientist_array):
    for scientist in scientist_array[0]:
        # Fields-capturing
        print(scientist)
        Name =  scientist.get("கண்டுபிடிப்பாளர்", None)
        birth = scientist.get("பிறப்பு",None)
        death = scientist.get("இறப்பு", None)
        place = scientist.get("வாழிடம்", None)
        nationality = scientist.get("தேசியம்", None)
        field = scientist.get("துறை", None)
        work_place = scientist.get("பணியிடங்கள்", None)
        education = scientist.get("கல்வி கற்ற இடங்கள்", None)
        prize = scientist.get("அறியப்படுவது",None)
        invention = scientist.get('கண்டுபிடிப்பு', None)

        yield {
            "_index": "scientists_db",
            "_source": {
                "கண்டுபிடிப்பாளர்": Name,
                "பிறப்பு": birth,
                "இறப்பு": death,
                "வாழிடம்": place,
                "தேசியம்": nationality,
                "துறை": field,
                "பணியிடங்கள்":work_place,
                "கல்வி கற்ற இடங்கள்": education,
                "அறியப்படுவது": prize,
                'கண்டுபிடிப்பு':  invention
            },
        }

# createIndex()
all_scientists = read_all_scientists()
helpers.bulk(client,genData(all_scientists))
```
